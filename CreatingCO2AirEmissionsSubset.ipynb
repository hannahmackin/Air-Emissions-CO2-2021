##Filtering


We could keep the df dataframe we just made as it is, but it contains excessive information that does not serve purpose of relevance to the intended research questions we might want to use. The subset we create will only include the following variables:


*   Reported State
*   Reported Latitude

##Filtering (continued)
To filter down to these variables, we need to use the "Series" feature of the dataframe to pull out certain desired columns from our tabular dataset, using either of these options:


*  dot notation `df.Region`
or
*  bracket notation `df["Region"]`.




We want to use the subset later on, so we will create it as follows using the `.copy()` method. I am calling my subset 'CO2_subset'.



```python
CO2_subset = df[["Reported State","Reported Latitude", "Reported Longitude", "CO2 emissions (non-biogenic) "]].copy()
```

### Exporting our New Subset

Now that our dataset has been narrowed down to be easier and more applicable to use for further analysis, we can export the subset as a new `.csv` file. This will make the dataset able to be shared with the public on our GitHub repositories.

#### Exporting to .csv file

We can use the code `.to_csv()` - filling in the filename and extension within the parentheses at the end.

For my filtered subset I would run: CO2_subset.to_csv("CO2_subset.csv")` this will export a `.csv` file in our working directory.

This `.csv` will automatically include the column of indices made by pandas when we linked the original file into our notebook using `.read_csv`.

We can easily remove these by including in our code `index=false`.
This will be our code:
`CO2_subset.to_csv("CO2_subset.csv", index=False)`

```python
CO2_subset.to_csv("CO2_subset.csv", index=False)
```
